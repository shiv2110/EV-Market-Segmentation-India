```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
data1 = pd.read_csv('EVIndia.csv')
data2 = pd.read_csv('EV_CARS _INDIA.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
for i in range(data1.shape[0]):
    data1.loc[i, 'Car'] = data1.loc[i, 'Car'].strip()
    data2.loc[i, 'Brand Name'] = data2.loc[i, 'Brand Name'].strip()
```

```python
data1.columns
```

```python
data2.columns
```

```python
data1.drop(['Transmission', 'VehicleType', 'PriceRange', 'Range', 'Capacity'], axis = 1, inplace = True)
```

```python
data1.head()
```

```python
data2.drop(['Transmission', 'No. of Airbags'], axis = 1, inplace = True)
```

```python
data2.head()
```

```python
data2['No. of Seats'].unique()
```

```python
data2.drop(['No. of Seats'], axis = 1, inplace = True)
```

```python
data2.head()
```

```python
print(data1.shape, data2.shape)
```

```python
data1['Car']
```

```python
data = pd.merge(left = data1, right = data2, 
                   left_on = ['Car'],
                   right_on = ['Brand Name'],
                   how = "left")
```

```python
data.drop(['Brand Name'], axis = 1, inplace = True)
```

```python
data.to_csv('mega_cars.csv', index = False)
```
